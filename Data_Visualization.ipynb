# Visualizing the data

In this module we will be visualising the dataset for better understanding of different activity classes and features associated with them.

This will help us to better understand how the changing features lead to distinction between various available activity classes.

```
# Imports
import numpy
from pandas import read_csv
import matplotlib
import matplotlib.pyplot as plt
```

It was easier to just read the data from raw file and then find the unique indices for each available class and then plot them.

```
# load a single file as a numpy array
def load_file(filepath, index):
	dataframe = read_csv(filepath, header=None, delim_whitespace=True)
	return dataframe.values[index]	
```

```
def plot_activity(activityIndex, figureSize=[12, 18]):
    # Define Parameters
    indices = [78, 150, 125, 27,  0,  51]
    labels = ['WALKING', 'WALKING UPSTAIRS', 'WALKING DOWNSTAIRS', 'SITTING', 'STANDING', 'LAYING']
    index = indices[activityIndex]
    label = labels[activityIndex]

    # Define filepaths
    fpath = '/content/drive/MyDrive/InnoHAR/Dataset/UCI_HAR_Dataset/train/Inertial Signals/'
    fileNames = ['body_acc_x_train.txt', 'body_acc_y_train.txt', 'body_acc_z_train.txt', 'body_gyro_x_train.txt', 'body_gyro_y_train.txt',
                 'body_gyro_z_train.txt', 'total_acc_x_train.txt', 'total_acc_y_train.txt', 'total_acc_z_train.txt']
    
    # Read Data
    data = []
    for fileName in fileNames:
        filePath = fpath + fileName
        loaded = load_file(filePath, index)
        data.append(loaded)

    # Define Graph Cycles
    titles = [
    'Body Acceleration - X', 'Body Acceleration - Y', 'Body Acceleration - Z', 
    'Body Gyro - X', 'Body Gyro - Y', 'Body Gyro - Z', 
    'Total Acceleration - X', 'Total Acceleration - Y', 'Total Acceleration - Z'
    ]


    # Define Graph parameters
    matplotlib.rcParams['figure.figsize'] = figureSize
    fig, axs = plt.subplots(3, 3)
    fig.suptitle('Different Features For Activity: ' + label, fontsize=16)

    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    # Plot Graphs
    index = 0
    for i in range(3):
        for j in range(3):
            axs[i][j].stem(data[index], use_line_collection=True)
            axs[i][j].set(ylabel=titles[index])
            index = index + 1
            
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])

#     # Hide x labels and tick labels for top plots and y ticks for right plots.
#     for ax in axs.flat:
#         ax.label_outer()
```

```
# Plotting Activity -> Walking
plot_activity(0)
```

```
# Plotting Activity -> Walking Upstairs
plot_activity(1)
```

```
# Plotting Activity -> Walking Downstairs
plot_activity(2)
```
